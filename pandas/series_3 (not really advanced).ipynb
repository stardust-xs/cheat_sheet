You can check if any value is present in the Series the same way you would check it if the value is in the list, using the **in** keyword.

For normal list:

```python
xa_list = ['XA', 'is', 'a', 'God']

'XA' in xa_list
```

```python
'You' in xa_list
```

Similarly, you can test it on Pandas series. But you need to specify that you are searching the value in values section. Pandas by default starts searching in the index of the series.

If you see below, it returns False when we ask if **Mewtwo** is in pokemon series or not. Of course, it is there. But pandas is searching **Mewtwo** in the pokemon's index list and not in the pokemon's values.

```python
'Mewtwo' in pokemon # Internally pandas is running >> 'Mewtwo' in pokemon.index
```

```python
'Mewtwo' in pokemon.values
```

## 8. Sorting in Series
You can sort either index or values in ascending or descending order using either **sort_index** and/or **sort_values** method.

By default, pandas sorts in **ascending** order.

```python
google.head(3)
```

As you can see, the series has already sorted the index in ascending order by default. But if you need to change their order of sorting, we can perform it using the **sort_index** method.

```python
google.sort_index(ascending=False).head(3) # Now we have sorted the index in descending order
```

If you need to sort the values in any order use, **sort_values** method.

```python
google.sort_values(ascending=True).head(3) # Ascending sort
```

```python
google.sort_values(ascending=False).head(3) # Descending sort
```

## 9. Inplace argument
When we apply some methods to the pandas series or dataframe object it creates a copy of it and works with that object leaving our main object intact. In simple terms, sort_index/sort_values methods have changed the state of **google** object but our main google object (unsorted) is still unchanged.

Below you can see, the main **google** object created at **In [13]** is the same after sorting. In a nutshell, most methods don't make a permanent change to the main object unless we specify it to do so.

If we want the changes to be permanent we need to specify **inplace=True** argument in the applied methods. Most of the Pandas method support **inplace** method.

```python
google.head() # Default google series object
```

```python
google.sort_values(ascending=False).head() # Sort in descending order
```

If you access google again, the sorting will be lost as we didn't make the sorting operation permanent. 

```python
google.head() # Sorting is lost
```

```python
google.sort_values(ascending=False, inplace=True)

google.head()
```

Inplace argument makes the changes permanent to the Pandas object. In case, the method doesn't support inplace argument we can assign the change to the same object. See below to understand:

Let me revert back to the original sorted series.

```python
google = pd.read_csv(filepath_or_buffer='csv/google.csv', squeeze=True)

google.head(3)
```

Now, we will apply **sort_values** method permanently to the google series **without using inplace=True**.

```python
google = google.sort_values(ascending=False)

google.head()
```

Here, we applied the **sort_values** method and re-assign it back to google. Power of Python!!

### 10. Extracting values from the Series index
Here, the pandas treats its Series object as a list. This is similar to using **iat** and **iloc** attributes. But simpler and more robust. We will use pokemon series for reference.

```python
pokemon.head(3)
```

If we want to find out the name/value of the Pokemon at position (index) 24 we can just treat the Series like a ordinary python list.

```python
pokemon[24]
```

This approach is similar to **iat** or **iloc**.

```python
pokemon.iat[24]
```

But the above method cannot retrieve multiple values. For retreiveing mulitple values you need to use the below approach.

```python
pokemon[[149, 247, 249, 383]] # All dope pokemons
```

Since, pandas series are treated like a list you can apply splicing too.

```python
pokemon[151:251] # All Gen2 pokemons
```

```python
pokemon[:151] # All Gen1 pokemons
```

#### Extracting values using index names/labels
We can extract the values using index (names/strings) too. Let me modify the Series little bit. Instead of using numbers as index, we would use the Pokemon names as the index - this can be done using parameter called **index_col**.

```python
pokemon = pd.read_csv(filepath_or_buffer='csv/pokemon.csv', index_col='Pokemon', squeeze=True)

pokemon.head()
```

Now we can try applying the same above approach of extracting values.

```python
pokemon['Mewtwo'] # Returns value of the Index (Mewtwo is the index and its value is Psychic)
```

```python
pokemon[['Larvitar', 'Pupitar', 'Tyranitar']]
```

```python
pokemon['Machop':'Machamp'] # Note that, when returning values using names/label the last values is also rendered
```

#### Extracting values using ".get()" method
As I said at the beginning pandas often treats its series as a list, but it also operates on it as a dictionary too. For instance, it uses the same **.get()** method.

Using **.get()** method is especially helpful when you don't know if the value exists. This method allows user to pass a default value when encountered with **NaN** or **None** value.

In the pokemon series, there is no Pokemon called as '**XAMES3**', if you try to call '**XAMES3**' all the above learnt approaches will throw a **KeyError** because we don't have any key to map in the Series.

```python
# pokemon['XAMES3'] 
# Throws TypeError and KeyError if executed

pokemon.get('XAMES3', default='Not a Pokémon') 
```

```python
pokemon.get('Mewtwo', default='Not a Pokémon')
```

#### If we pass a list of elements, the default value will be overwritten by "NaN"

```python
pokemon.get(['Mewtwo', 'Tyranitar', 'Rayquaza', 'XAMES3'], default='Not a Pokémon')
```
